```python
import sqlite3
conn = sqlite3.connect("myData2.db")   # 資料庫連線
cursor = conn.cursor()
sql = '''Create table students(  
        id int,
        name text,
        gender text)'''
cursor.execute(sql)                     # 執行SQL指令
cursor.close()                          # 關閉
conn.close()                            # 關閉資料庫連線
```

# Insert Data

```python
import sqlite3
conn = sqlite3.connect("myData3.db")     # 資料庫連線
print("請輸入myData3資料庫students表單資料")
while True:
    new_id = int(input("請輸入id : "))  # 轉成整數
    new_name = input("請輸入name : ")
    new_gender = input("請輸入gender : ")
    x = (new_id, new_name, new_gender)
    sql = '''insert into students values(?,?,?)'''  
    conn.execute(sql,x)
    conn.commit()                       # 更新資料庫
    again = input("繼續(y/n)? ")
    if again[0].lower() == "n":
        break
conn.close()                            # 關閉資料庫連線
```

# Query Data

```python
import sqlite3
conn = sqlite3.connect("myData3.db")     # 資料庫連線
results = conn.execute("SELECT * from students")
for record in results:
    print("id = ", record[0])
    print("name = ", record[1])
    print("gender = ", record[2])
conn.close()                            # 關閉資料庫連線
```

```python
import sqlite3
conn = sqlite3.connect("myData3.db")     # 資料庫連線
results = conn.execute("SELECT * from students")
allstudents = results.fetchall()        
# 結果轉成元素是元組的串列

for student in allstudents:
    print(student)
conn.close()                            # 關閉資料庫連線
```

```python
import sqlite3
conn = sqlite3.connect("myData3.db")     # 資料庫連線
sql = '''SELECT name, gender
        from students
        where gender = "f"'''
results = conn.execute(sql)
allstudents = results.fetchall()        # 結果轉成元素是元組的串列
for student in allstudents:
    print(student)
conn.close()                            # 關閉資料庫連線
```

# Update Data

```python
import sqlite3
conn = sqlite3.connect("myData3.db")     # 資料庫連線
sql = '''UPDATE students
        set name = "Tomy"
        where id = 1'''
results = conn.execute(sql)
conn.commit()                           # 更新資料庫
results = conn.execute("SELECT name from students")
allstudents = results.fetchall()        
# 結果轉成元素是元組的串列
for student in allstudents:
    print(student)
conn.close()                            # 關閉資料庫連線
```

# Delete Data

```python
import sqlite3
conn = sqlite3.connect("myData3.db")     # 資料庫連線
sql = '''DELETE
        from students
        where id = 2'''
results = conn.execute(sql)
conn.commit()                           # 更新資料庫
results = conn.execute("SELECT name from students")
allstudents = results.fetchall()        # 結果轉成元素是元組的串列
for student in allstudents:
    print(student)
conn.close()                            # 關閉資料庫連線
```
